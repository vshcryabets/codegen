# Trees

1. **AST (Abstract Syntax Tree)** - The initial tree structure generated by parsing and/or executing the description files.
1. **Syntax Parse Tree** - This tree is the result of applying transformation logic to the **AST** (see BuildOutTreeUseCase).
1. **Stylistic Tree** - The final tree, which is produced by applying code style rules to the **Syntax Parse Tree** (see PrepareCodeStyleTreeUseCase).
